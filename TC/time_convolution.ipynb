```python
import numpy as np

import torch
import torch.optim as optim
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, TensorDataset, DataLoader
from torch.utils.data.dataset import random_split
print(torch.cuda.is_available())
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(device)

import matplotlib.pyplot as plt
%matplotlib inline
torch.manual_seed(42)
```

```python
# Load data
loaded_trajectory_pos = np.load('trajectory_pos.npy', allow_pickle=True)
N_STEPS = len(loaded_trajectory_pos)
NUM_PARTICLES = loaded_trajectory_pos[0].shape[0]
LX = 5. # BOX WIDTH
LY = 5.  # BOX HEIGHT
DEG_FREED = 2 # degrees of freedom, here x and y coordinates
NUM_FEATURES = NUM_PARTICLES * DEG_FREED
print('N_STEPS =', N_STEPS)
print('NUM_PARTICLES =', NUM_PARTICLES)
print('DEG_FREED =', DEG_FREED)
print(loaded_trajectory_pos.shape)
```

```python
# Data preparation
#input T0 is the time step t; output T1 the time step t+1
sequence = loaded_trajectory_pos/LX
#all values are betwenn 0 and 1
print(sequence.shape)
print(np.max(sequence))
print(np.min(sequence))
```

```python
#Data preparation
sequence_tensor = torch.tensor(sequence, dtype=torch.float32)
print(sequence_tensor.shape)
```

```python
P_STEPS = 3  # Use previous P_STEPS steps for prediction

def create_batches(data, batch_size):
    # Ensure that the number of batches is such that the last batch has at least P_STEPS time steps
    num_batches = (len(data) - P_STEPS + 1) // batch_size
    batches = [data[i*batch_size:(i+1)*batch_size + P_STEPS - 1] for i in range(num_batches)]
    return batches

# Create batches
batch_size = 16  # Define batch size
batches = create_batches(sequence_tensor, batch_size)
```

The `create_batches` function is designed to create overlapping batches from the `data` sequence for training a model on sequential data. Let's break down the function and understand the contents of batches like `batch[0]`, `batch[1]`, etc.

1. **Function Overview**:
   - `data` is your sequence, with each element (e.g., `x0`, `x1`, ..., `x1001`) representing a timestep in the sequence. Each timestep has a shape of `[12, 2]`.
   - `batch_size` is the number of sequences you want in each batch.
   - `P_STEPS` is the number of timesteps considered for predicting the next timestep.

2. **Calculating `num_batches`**:
   - The total number of batches (`num_batches`) is calculated based on the length of `data`, `batch_size`, and `P_STEPS`.
   - `(len(data) - P_STEPS + 1)` ensures that each sequence in the last batch has enough preceding timesteps for prediction.

3. **Creating Batches**:
   - Batches are created such that each batch contains sequences overlapping with the next batch.
   - Each sequence in a batch consists of `P_STEPS` timesteps.

4. **Contents of `batch[0]`, `batch[1]`, etc.**:
   - `batch[0]`: Contains the first `batch_size + P_STEPS - 1` elements from `data`. For example, if `batch_size` is 50 and `P_STEPS` is 3, `batch[0]` will contain elements `x0` to `x51` (52 elements in total).
   - `batch[1]`: Starts from the element at index `batch_size` in `data` and includes the next `batch_size + P_STEPS - 1` elements. Continuing the example, `batch[1]` will contain elements `x50` to `x101`.
   - Subsequent batches follow the same pattern.

5. **Overlap Between Batches**:
   - There is an overlap of `P_STEPS - 1` elements between consecutive batches. This overlap ensures that each sequence within a batch has `P_STEPS` preceding timesteps available for prediction.

6. **Example**:
   - If `batch_size` is 50, `P_STEPS` is 3, and `data` contains 1001 elements (`x0` to `x1000`), then:
     - `batch[0]` will contain `x0` to `x51`.
     - `batch[1]` will contain `x50` to `x101`.
     - The last batch will include the last `batch_size + P_STEPS - 1` elements of `data`.

In summary, each batch contains a sequence of overlapping elements from `data`, where the overlap is determined by `P_STEPS`. This setup is typical in sequence prediction tasks where a fixed number of previous timesteps (`P_STEPS`) are needed to predict the next timestep.

```python
from torch.utils.data import DataLoader, Subset

# Shuffle and split indices for training and validation sets
batch_indices = list(range(len(batches)))
np.random.shuffle(batch_indices)
train_indices = batch_indices[:int(0.8 * len(batch_indices))]
val_indices = batch_indices[int(0.8 * len(batch_indices)):]

# Construct training and validation datasets
train_data = torch.cat([batches[i] for i in train_indices], dim=0)
validation_data = torch.cat([batches[i] for i in val_indices], dim=0)

# DataLoader for training and validation sets
train_loader = DataLoader(train_data, batch_size=batch_size)
validation_loader = DataLoader(validation_data, batch_size=batch_size)

print(train_data.shape)
print(validation_data.shape)
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class ConvolutionalModel(nn.Module):
    def __init__(self, n_particles, d_freedom, hidden_dim, p_steps):
        super(ConvolutionalModel, self).__init__()
        self.particles = n_particles
        self.degrees = d_freedom
        self.p_steps = p_steps
        self.features = n_particles * d_freedom
        self.hidden_dim = hidden_dim

        # Convolutional layer with kernel_size equal to p_steps to mix information across all time steps
        # No padding is used, so the convolution spans the entire input sequence
        self.conv = nn.Conv1d(in_channels=self.features, out_channels=hidden_dim, kernel_size=p_steps, padding=0)

        # Final linear layer to map the convolution output to the desired output size
        # Since the convolution output will be a single value per feature channel, the input size is hidden_dim
        self.final_layer = nn.Linear(hidden_dim, self.features)

    def forward(self, x):
        # Input x shape: [batch_size, P_steps, N_particles, degrees of freedom]
        # Reshape x to [batch_size, P_steps, n_features] where n_features = N_particles * degrees of freedom
        batch_size, _, _, _ = x.size()
        x = x.view(batch_size, self.p_steps, self.features)
        # Permute x to [batch_size, n_features, P_steps] to align with Conv1d input shape [batch, channels, length]
        # Here, n_features are treated as channels by the convolution operation
        x = x.permute(0, 2, 1)
 
        # Apply the 1D convolution
        # The convolution mixes information across all time steps for each feature channel
        x = F.relu(self.conv(x))
 
        # After convolution, the output shape is [batch_size, hidden_dim, 1] due to kernel_size=p_steps and padding=0
        # Squeeze the last dimension to make it [batch_size, hidden_dim] for the linear layer
        x = x.squeeze(-1)
 
        # Apply the final linear layer
        x = self.final_layer(x)
 
        # Reshape the output to match the original input shape, if necessary
        # The output shape is now [batch_size, n_particles, degrees of freedom]
        output = x.view(batch_size, self.particles, self.degrees)
        return output
```

```python
N_HIDDEN = 24
# Create the model and test it
torch.manual_seed(42)
i=P_STEPS+1
try_input = torch.cat([batches[0][i-P_STEPS:i]], dim=0).unsqueeze(0)
print("input", try_input.shape)
model = ConvolutionalModel(NUM_PARTICLES, DEG_FREED, N_HIDDEN, P_STEPS).to(device)
out= model(try_input.to(device))
print("out shape", out.shape)
```

```python
# Define hyperparameters
learning_rate = 0.001
# Define loss function and optimizer
loss_fn = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=learning_rate)
```

```python
i=100
input_seq = torch.cat([train_data[i-P_STEPS:i]], dim=0)
print(input_seq.shape)
```

```python
# Training loop
num_epochs = 200
training_losses = []
validation_losses = []

for epoch in range(num_epochs):
    model.train()  # Set the model to training mode
    total_train_loss = 0.

    for batch in train_loader:
        # Move the batch to the device
        batch = batch.to(device)

        # Zero the gradients at the start of a new batch
        optimizer.zero_grad()

        # Iterate through each time step in the batch, starting from P_STEPS
        for i in range(P_STEPS, len(batch)):
            # Prepare the input and target data for this time step
            input_seq = torch.cat([batch[i-P_STEPS:i]], dim=0).unsqueeze(0)
            target_seq = batch[i].unsqueeze(0)

            # Forward pass
            predicted_seq = model(input_seq)

            # Compute the loss
            loss = loss_fn(predicted_seq, target_seq)
            total_train_loss += loss.item()

            # Accumulate gradients
            loss.backward()

        # Optimization step after processing the entire batch
        optimizer.step()

    # Append loss
    training_losses.append(total_train_loss)

    # Validation loop
    model.eval()  # Set the model to evaluation mode
    total_val_loss = 0.
    with torch.no_grad():
        for batch in validation_loader:
            # Move the batch to the device
            batch = batch.to(device)

            # Iterate through each time step in the batch
            for i in range(P_STEPS, len(batch)):
                # Prepare input and target sequences from validation batch
                input_seq = torch.cat([batch[i-P_STEPS:i]], dim=0).unsqueeze(0)
                target_seq = batch[i].unsqueeze(0)

                # Forward pass
                predicted_seq = model(input_seq)

                # Compute the loss
                val_loss = loss_fn(predicted_seq, target_seq)
                total_val_loss += 5.*val_loss.item()

    #Append loss
    validation_losses.append(total_val_loss)

    # Print losses for monitoring
    if epoch % ((num_epochs+1)//10) == 0:
        print(f"Epoch: {epoch} | Train loss: {total_train_loss:.4f} | Val loss: {total_val_loss:.4f}")
```

```python
plt.plot(training_losses, label='Training')
plt.plot(validation_losses, label='Validation')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training and Validation Losses')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

def make_video(real_y, pred_y, file_name):
    # Update function for the animation
    def update(frame_number, real_y, pred_y, scatter_real, scatter_pred, time_text):
        # Update the data of both scatter plots for the new frame
        scatter_real.set_offsets(real_y[frame_number])
        scatter_pred.set_offsets(pred_y[frame_number])
        # Update the time annotation
        time_text.set_text(f'Time = {frame_number}')
    
    # Set up the figure, the axis, and the plot elements we want to animate
    fig, ax = plt.subplots()
    ax.set_aspect('equal')
    ax.set_xlim([-0.1, 1.1])
    ax.set_ylim([-0.1, 1.1])
    
    # Create initial scatter plots and a text annotation
    scatter_real = ax.scatter(real_y[0, :, 0], real_y[0, :, 1], color='blue', marker='o')
    scatter_pred = ax.scatter(pred_y[0, :, 0], pred_y[0, :, 1], color='orange', marker='*')
    time_text = ax.text(0.5, 1.05, '', transform=ax.transAxes, ha='center')
    
    # Create the animation using FuncAnimation
    ani = FuncAnimation(fig, update, frames=len(real_y), fargs=(real_y, pred_y, scatter_real, scatter_pred, time_text), interval=100)
    
    # Save the animation
    ani.save(file_name, writer='ffmpeg')
    
    # Convert the animation to an HTML5 video embedded in the notebook
    html = HTML(ani.to_html5_video())
    
    # Close the figure to prevent it from displaying in the notebook or script output
    plt.close(fig)
    
    # Return the HTML object to display the animation in the notebook
    return html
```
